```python
import numpy as np
import matplotlib.pyplot as plt

deger1 = np.random.randint(100, 250, 5)
deger2 = np.random.randint(100, 250, 5)

dizi_ismi = ["Bitcoin", "Ethereum"]

ort_btc = np.mean(deger1)
ort_eth = np.mean(deger2)

new_dizi = [ort_btc, ort_eth]

plt.bar(dizi_ismi, new_dizi, color=["orange", "green"], label=["BTC", "ETH"])

plt.legend()

for i in range(len(dizi_ismi)):
    plt.text(i, new_dizi[i], f"{new_dizi[i]:.2f}", ha="center", va="bottom")

plt.xticks(rotation=45)  # x eksenindeki etiketleri 45 derece dndrme

plt.show()
```
